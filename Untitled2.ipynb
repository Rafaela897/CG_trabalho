### Gerar csv para a previsão das 4 semanas seguinte para todas as lojas e departamentos

```python
filename = 'model.sav'
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
num = df['Date'].tail(1).values
start_date = datetime.datetime.strptime(num[0],'%Y-%m-%d')

date = []
store = []
dept = []
holiday = []
year = []
month=[]
day=[]

for i in range(0,4):
  end_date = start_date + datetime.timedelta(days=7)

  new_date = end_date.strftime('%Y-%m-%d')
  for store_num in range(1,df['Store'].max()+1):
    dept_aux = df.query('Store=='+str(store_num))
    for dept_num in pd.unique(dept_aux['Dept']):
      date.append(new_date)
      day.append(getDay(new_date))
      month.append(getMonth(new_date))
      year.append(getYear(new_date))
      store.append(store_num)
      dept.append(dept_num)
      holiday.append('False')

  start_date = end_date

#holiday[-1]='True'
print(holiday[-1])
newdataset = pd.DataFrame(list(zip(date,store,dept,holiday, year, month, day)), columns = [ 'Date','Store', 'Dept', 'IsHoliday', 'Year', 'Month', 'Day'])
print(newdataset)
```
